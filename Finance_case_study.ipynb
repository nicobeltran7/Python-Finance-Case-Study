```python
import yfinance as yf

# Download stock data for AAPL
ticker = "AAPL"
stock = yf.download(ticker, start="2021-01-01", end="2024-01-31")

# Display the DataFrame
print(stock.head())
```

```python
stock.columns = stock.columns.droplevel(1)  # Drop the first level of columns
```

```python
stock
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="Date", y="Close", data=stock)
plt.title("AAPL stock price")
plt.ylabel("Price in USD")
```

```python
stock["20day"] = stock["Close"].rolling(window=20).mean()
stock["50day"] = stock["Close"].rolling(window=50).mean()
stock
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="Date", y="Close", data=stock, label="Close Price")
sns.lineplot(x="Date", y="20day", data=stock, label="20 day moving")
sns.lineplot(x="Date", y="50day", data=stock, label="50 day moving")
plt.title("AAPL stock price")
plt.ylabel("Price in USD")
```

```python
stock["Daily return(%)"] = stock["Close"].pct_change()
stock["Daily std dev"] = stock["Daily return(%)"].rolling(window=20).std()
stock
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="Date", y="Daily std dev", data=stock, label="Close Price")

plt.title("Rolling 20 day volatility")
plt.ylabel("Volatility")
```

```python
daily_volatitlity = stock["Daily return(%)"].std()
daily_volatitlity
```

```python
annualized_volatility = daily_volatitlity*(252**0.5)
annualized_volatility
```

```python
print(f"Daily Volatility:{daily_volatitlity}")
print(f"Annualized Volatility:{annualized_volatility}")
```

```python
tickers = ["GOOGL","AAPL","JPM","AMZN","^GSPC"]
stock_data= yf.download(tickers, start = "2021-01-01", end="2024-01-31")["Close"]
stock_data
```

```python
stock_returns = stock_data.pct_change()*100
stock_returns
```

```python
stock_returns = stock_data.pct_change()
stock_returns
stock_returns_smooth = stock_returns.rolling(window=100).mean()

plt.figure(figsize=(14, 7))
for ticker in tickers:  # Loop through the list of tickers
    sns.lineplot(data=stock_returns_smooth[ticker], label=ticker)
plt.title("Smooth 100 average returns comparison: XOM, AAPL, JPM, KO")
plt.ylabel("Daily returns(%)")
```

```python
plt.figure(figsize=(14, 7))
for ticker in tickers:  # Loop through the list of tickers
    sns.lineplot(data=stock_data[ticker], label=ticker)
plt.title("Share price trend comparison")
plt.ylabel("Share Price")
```

```python
tickers_corr = ["GOOGL","AAPL","AMZN","KO","^GSPC"]
stock_data_corr= yf.download(tickers, start = "2021-01-01", end="2024-01-31")["Close"]
stock_data_corr

stocks_returns_corr = stock_data_corr.pct_change()*100
correlation_matrix = stocks_returns_corr.corr()
correlation_matrix
```
