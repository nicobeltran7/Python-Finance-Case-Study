```python
ticker = "AAPL"  # if empty, try "AAPL.US"
stock = web.DataReader(ticker, "stooq",start="2021-01-01",end="2024-01-31").sort_index()
stock.tail(20)
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="Date", y="Close", data=stock)
plt.title("AAPL stock price")
plt.ylabel("Price in USD")
```

```python
#Why 20/50-day moving averages?
#I plot Close plus 20-day (short-term trend) and 50-day (medium-term trend) SMAs.

#They smooth noise and reveal regime shifts without overfitting.

#Crossovers arenâ€™t used as trading signals here; they provide context for later risk metrics and multi-asset comparison.
```

```python
stock["20day"] = stock["Close"].rolling(window=20).mean()
stock["50day"] = stock["Close"].rolling(window=50).mean()
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="Date", y="Close", data=stock, label="Close Price")
sns.lineplot(x="Date", y="20day", data=stock, label="20 day moving")
sns.lineplot(x="Date", y="50day", data=stock, label="50 day moving")
plt.title("AAPL stock price")
plt.ylabel("Price in USD")
```

```python
stock["Daily return(%)"] = stock["Close"].pct_change()
stock["Daily std dev"] = stock["Daily return(%)"].rolling(window=20).std()
stock.tail(20)
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(x="Date", y="Daily std dev", data=stock, label="Close Price")

plt.title("Rolling 20 day volatility")
plt.ylabel("Volatility")
```

```python
#Volatility methodology

#Compute daily returns as % change of Close.

#Rolling 20-day stdev â‰ˆ 1-month realized volatility to visualize near-term risk.

#Annualized vol = daily stdev Ã— âˆš252 (trading days).
#This converts raw day-to-day fluctuations into an interpretable risk number for comparing assets and periods.
```

```python
daily_volatitlity = stock["Daily return(%)"].std()
daily_volatitlity
```

```python
annualized_volatility = daily_volatitlity*(252**0.5)
annualized_volatility
```

```python
print(f"Daily Volatility:{daily_volatitlity}")
print(f"Annualized Volatility:{annualized_volatility}")
```

```python
tickers = ["GOOGL","AAPL","JPM","AMZN","^GSPC"]
start, end = "2021-01-01", "2024-01-31"

# Stooq uses .US for U.S. equities and different codes for some indices.
index_map = {"^GSPC": "^SPX"}  # S&P 500 on Stooq
def to_stooq(t):
    if t in index_map:
        return index_map[t]
    return t if t.startswith("^") else f"{t}.US"

stooq_syms = {t: to_stooq(t) for t in tickers}

# Fetch each tickerâ€™s Close and combine
cols = []
for orig, stq in stooq_syms.items():
    s = web.DataReader(stq, "stooq", start=start, end=end)["Close"].sort_index()
    s.name = orig
    cols.append(s)

stock_data = pd.concat(cols, axis=1)

# show the most recent 20 rows
stock_data.tail(20)
```

```python
stock_returns = stock_data.pct_change()*100
stock_returns.tail(20)
```

```python
stock_returns = stock_data.pct_change()
stock_returns
stock_returns_smooth = stock_returns.rolling(window=100).mean()

plt.figure(figsize=(14, 7))
for ticker in tickers:  # Loop through the list of tickers
    sns.lineplot(data=stock_returns_smooth[ticker], label=ticker)
plt.title("Smooth 100 average returns comparison: XOM, AAPL, JPM, KO")
plt.ylabel("Daily returns(%)")
```

```python
plt.figure(figsize=(14, 7))
for ticker in tickers:  # Loop through the list of tickers
    sns.lineplot(data=stock_data[ticker], label=ticker)
plt.title("Share price trend comparison")
plt.ylabel("Share Price")
```

```python
tickers_corr = ["GOOGL", "AAPL", "AMZN", "KO", "^GSPC"]
start, end = "2021-01-01", "2024-01-31"

# Map your symbols to Stooq symbols
index_map = {"^GSPC": "^SPX"}  # S&P 500 on Stooq
def to_stooq(t):
    return index_map.get(t, t if t.startswith("^") else f"{t}.US")

# Fetch Close for each and combine
series_list = []
for t in tickers_corr:
    s = web.DataReader(to_stooq(t), "stooq", start=start, end=end)["Close"].sort_index()
    s.name = t
    series_list.append(s)

stock_data_corr = pd.concat(series_list, axis=1)

# Returns and correlation
stocks_returns_corr = stock_data_corr.pct_change() * 100
correlation_matrix = stocks_returns_corr.corr()

correlation_matrix.round(3)
```

```python
#Cross-asset story & diversification
# Normalize each ticker to returns (not prices) and map symbols to Stooq codes (e.g., ^GSPC â†’ ^SPX) to ensure apples-to-apples comparisons.

#The 100-day smoothed return chart shows broad regimes across names.

#The correlation matrix highlights co-movement: mega-cap tech names cluster together, while staples/financials (e.g., KO, JPM) show lower correlation to tech and the S&P 500â€”useful when reasoning about diversification and portfolio construction.
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap="Blues")
plt.title("Correlation stocks")
```
